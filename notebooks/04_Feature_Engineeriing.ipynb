```python
import os
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/telco_customer_churn_cleaned.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
print("Dataset Shape:", df.shape)
```

```python
df.isnull().sum()
```

```python
df["Churn"].value_counts()
```

```python
X = df.drop(columns=["customerID", "Churn"])

y = df["Churn"]
```

```python
print("Feature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
X.columns
```

```python
target_encoder = LabelEncoder()
```

```python
y = target_encoder.fit_transform(y)
```

```python
print("Unique values:", np.unique(y))

print("\nFirst 10 values:")

print(y[:10])
```

```python
categorical_columns = X.select_dtypes(include=["object", "string"]).columns

print("Categorical Columns:")
print(categorical_columns)
```

```python
print("Number of categorical columns:", len(categorical_columns))
```

```python
X = pd.get_dummies(
    X,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)
```

```python
print("Encoded Dataset Shape:", X.shape)
```

```python
X.columns.tolist()
```

```python
X.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print()

print("Training Labels   :", y_train.shape)
print("Testing Labels    :", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_columns = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Fit ONLY on training data
X_train_scaled[numerical_columns] = scaler.fit_transform(
    X_train[numerical_columns]
)

# Transform test data using the SAME scaler
X_test_scaled[numerical_columns] = scaler.transform(
    X_test[numerical_columns]
)
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)
```

```python
train_data = X_train_scaled.copy()
train_data["Churn"] = y_train

train_data.to_csv(
    "../data/processed/train_scaled.csv",
    index=False
)
```

```python
test_data = X_test_scaled.copy()
test_data["Churn"] = y_test

test_data.to_csv(
    "../data/processed/test_scaled.csv",
    index=False
)
```

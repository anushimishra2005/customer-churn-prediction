```python
categorical_columns = df.select_dtypes(include="object").columns.tolist()

categorical_columns
```

```python
def analyze_categorical_feature(feature):
    
    plt.figure(figsize=(8,5))

    sns.countplot(
        data=df,
        x=feature,
        hue="Churn"
    )

    plt.title(f"Customer Churn by {feature}")

    plt.xticks(rotation=20)

    plt.show()

    print("="*60)

    print(f"Percentage Distribution for {feature}")

    print()

    percentage = pd.crosstab(
        df[feature],
        df["Churn"],
        normalize="index"
    )*100

    display(percentage.round(2))
```

Key Insights from Exploratory Data Analysis

## 1. Customer Churn Distribution
- Around 27% of customers churned.
- The dataset is slightly imbalanced.

## 2. Tenure
- Customers with shorter tenure are more likely to churn.

## 3. Monthly Charges
- Churned customers pay higher monthly charges on average.

## 4. Total Charges
- Customers who stay generate much higher lifetime revenue.

## 5. Contract Type
- Month-to-month contracts have the highest churn.
- Two-year contracts have the lowest churn.

## 6. Internet Service
- Fiber optic customers show higher churn.

## 7. Tech Support & Online Security
- Customers without these services churn more frequently.

## 8. Demographics
- Gender has little effect on churn.
- Senior citizens churn more often.

## Overall Business Recommendation
Focus retention efforts on:
- New customers
- Month-to-month contracts
- Fiber optic users
- High monthly charge customers
- Customers without tech support or online security
